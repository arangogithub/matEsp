# Heat maps y sucesiones de Cauchy
## Un método experimental para determinar convergencia

Dada una sucesión de números reales $(x_n)$, generaremos un heat map de las distancias 
$ 
\left|{x_n-x_m}\right|
$. Primero creamos una matriz cuadrada $N\times N$, luego visualizamos la matriz con alguna convención de colores (cmap). Este notebook ilustra el caso

$$
x_n = \sin{n}
$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nsteps = 1300
data = np.array([np.abs(np.sin(n)-np.sin(m)) for n in range(Nsteps) for m in range(Nsteps)]) 
```

```python
matrix = np.array(data.reshape(Nsteps,Nsteps))
```

```python
plt.matshow(matrix, fignum=None, cmap= 'hot')
```
